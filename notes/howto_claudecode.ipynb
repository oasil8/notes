claude는 코드베이스 전체랑 git 커밋 히스토리에 접근할 수 있다. 필요한 정보를 얻기 위해 bash(터미널셀) 명령어도 실행가능함. 

.claude 폴더 안에 settings.loal.json으로 저장하면, 해당 프로젝트에만 적용되고 git에는 포함되지 않음.
같은 위치지만 'local' 없이 저장하면, 프로젝트 전체에 적용되고 git에 커밋돼서 팀이랑 공유됨.
홈 디렉터리 ~/.claude/settings.json에 저장하면, 모든 프로젝트에 공통으로 적용됨.

<claude code의 3가지 모드> 
각각 언제 쓰는지 알면 생산성이 훨씬 올라갈거임.
모드를 바꾸고 싶으면 Shift + Tab 누르면 됨.

1. 기본(default) 모드: claude code가 파일을 수정할 때마다 내 허락을 먼저 구하는 모드
2. 자동허용(auto-accept) 모드: 수정할 때마다 물어보는 게 귀찮을 때 유용함.
3. 계획(plan) 모드: claude가 무작정 코드를 작성하지 않고 일단 문제를 분석하도록 할 때 쓰면 됨.
		  	먼저 문제를 차근차근 생각해보고, 가능한 해결 방법들을 떠올리고, 계획을 세운 다음에 코드를 짜게 만드는 거임.
			복잡한 기능을 구현하거나 구현 방식을 확신 못할 때 특히 유용함.
			이 기능 덕분에 claude가 내가 의도한거랑 완전 다른 걸 만들 뻔한 상황을 수없이 피할 수 있었음-노마드 코더

여러 메시지를 한꺼번에 예약해두면 claude가 순서대로 처리해줌.
    예를 들면 하나 요청하고 claude가 처리하는 동안 또 다른 거를 계속 추가할 수 있음.

claude가 가끔 이상한 방향으로 갈때 프로그래머라면 ctrl + c 를 누르는데 -> 절대 그러지 말것. 그러면 프로그램이 죽어버림.
대신 Esc를 누르자. 한번 누르면 claude를 멈추고, 두번 누르면 이전 메시지로 돌아가서 대화를 포크할 수 있음.

claude code는 터미널 안에서 작동하니깐 내 시스템에 있는 어떤 cli 도구도 쓸 수 있음. (사용법만 알면 말이지.)
    예를 들어, claude한테 playwright 쓰는 법을 알려주고 싶다면 playwright --help 명령어를 실행하면 됨. 
    이건 UI 만들때 완전 유용함. -> 내가 claude한테 UI 디자인 이미지를 주면 cladue가 ui를 만들고 playwright로 스크린샷을 찍어서 이미지랑 딱 맞을 때까지 반복함.
    즉, claude code한테 자기 작업을 확인할 수 있는 도구만 주면, 그것이 스크린샷이든 테스트이든 무엇이든간에 그걸 이용해서 계속 반복하고 수정해서 원하는 결과가 나올때까지 작업함.

추가적으로 더 좋은 경험을 원한다면, cladue code for vscode 익스텐션도 설치해보자.
    클릭 한번으로 claude code를 새 패널에서 띄울 수 있고, 코드 차이(diff)는 터미널 대신 에디터에서 보이고, 코드 블록을 선택하면 claude가 그걸 바로 인식함.
